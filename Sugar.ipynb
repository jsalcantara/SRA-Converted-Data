```python
import pickle as pkl
import pandas as pd

# List of pickle files
pkl_files = [
    "annual_area_harvested.pkl",
    "annual_ave_hours_grinding.pkl",
    "annual_cane_production.pkl",
    "annual_capacity_utilization.pkl",
    "annual_mechanical_efficiency.pkl",
    "annual_rated_capacity.pkl",
    "annual_recovery.pkl",
    "annual_tons_cane_hour.pkl"
]

# Convert each pickle file to CSV
for pkl_file in pkl_files:
    # Load pickle file
    with open(pkl_file, "rb") as f:
        data = pkl.load(f)
    
    # Convert to DataFrame
    df = pd.DataFrame(data)
    
    # Create CSV file name by replacing .pkl with .csv
    csv_file = pkl_file.replace(".pkl", ".csv")
    
    # Save as CSV
    df.to_csv(csv_file, index=False)

print("Conversion complete! All files saved as CSV.")
```
